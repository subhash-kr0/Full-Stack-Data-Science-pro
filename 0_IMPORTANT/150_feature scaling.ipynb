```python
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#scaling (optional)>>brings oall the feature on the same scale
```

```python
#why scaling? --- To reduce the computation time(faster optimisation) and better interpretation
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
#standardisation >> zsore = x-mu/sigma
mean = np.mean(df['total_bill'])
```

```python
std = np.std(df['total_bill'])
```

```python
std_data = []
for i in list(df['total_bill']):
    zscore = (i-mean)/std
    std_data.append(zscore)
```

```python
std_data
```

```python
#scaling is optional because it doesnt change the distribution of data
#before scaling
sns.distplot(df['total_bill'])
```

```python
#after scaling
sns.distplot(std_data)
```

```python
scaler = StandardScaler()
```

```python
scaler
```

```python
scaler.fit(df[['total_bill']]) #it will calculate mu and sigma
```

```python
scaler.transform(df[['total_bill']])
```
