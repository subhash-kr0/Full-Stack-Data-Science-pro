2. Given a recurrence relation, solve it using the substitution method:

a. T(n) = T(n-1) + c

b. T(n) = 2T(n/2) + n

c. T(n) = 2T(n/2) + c

d. T(n) = T(n/2) + c

Sol:- (a) T(n) = T(n-1) + c

   T(n) = T(n-2) + 2c  ------2nd time

   T(n) = T(n-3) + 3c  ------3rd time

   lets assume it continue k-times, n-k = 1

  T(n) = T(n-k) + kc

  => T(n) = T(n - (n-1)) + (n-1)c

  => T(n) = O(n) , Linear time complexity

<br> (b) T(n) = 2T(n/2) + n
<br> T(n) = 2T(n/2) + n

<br> T(n) = 4T(n/4) + 2(n/2) + n
<br> T(n) = 4T(n/4) + n + n
<br> T(n) = 4T(n/4) + 2n
<br>
<br> Continuing this process:
<br> T(n) = 2^k T(n/2^k) + k n
<br>
<br> When n/2^k = 1, n = 2^k, k = log_2(n)
<br>
<br> Substitute k:
<br> T(n) = 2^log_2(n) T(1) + n log_2(n)
<br> T(n) = n T(1) + n log_2(n)
<br>
<br> Assuming T(1) = c:
<br> T(n) = n c + n log_2(n)
<br> T(n) = O(n log n)

<br> Continuing this process:
<br> T(n) = 2^k T(n/2^k) + k c
<br>
<br> When n/2^k = 1, n = 2^k, k = log_2(n)
<br>
<br> Substitute k:
<br> T(n) = 2^log_2(n) T(1) + log_2(n) c
<br> T(n) = n T(1) + log_2(n) c

<br> Assuming T(1) = d:
<br> T(n) = n d + c log_2(n)

<br> T(n) = T(n/2) + c

<br> T(n) = T(n/4) + 2c

<br> T(n) = T(n/2^k) + k c
<br>
<br> When n/2^k = 1, n = 2^k, k = log_2(n)
<br>
<br> Substitute k:
<br> T(n) = T(1) + log_2(n) c
<br>
<br> Assuming T(1) = d:
<br> T(n) = d + c log_2(n)
<br> T(n) = O(log n)

3. Given a recurrence relation, solve it using the recursive tree approach:<br>
a. T(n) = 2T(n-1) +1<br>
b. T(n) = 2T(n/2) + n

<br> (a) T(n) = 2T(n-1) + 1<br>
<br> Visualize the Recursion Tree:
<br> - The root node represents T(n).
<br> - The root splits into two nodes, each representing T(n-1), plus 1 for the root level work.
<br> - Each T(n-1) further splits into two T(n-2) nodes, and so on.<br>
<br> Work Done at Each Level:
<br> - Level 0: T(n)
<br> - Level 1: 2T(n-1) + 1
<br> - Level 2: 4T(n-2) + 2
<br> - Level k: 2^k T(n-k) + (2^k - 1)<br>
<br> Summing the Work:
<br> - The number of nodes at level k is 2^k.
<br> - The total work at level k is 2^k T(n-k).
<br> - The total work for all levels is the sum of the work at each level.<br>
<br> Base Case:
<br> - T(0) is a constant.<br>
<br> Generalize:
<br> - T(n) = 2^0 T(n-0) + 2^1 T(n-1) + 2^2 T(n-2) + ... + 2^n T(0)
<br> - Simplify to T(n) = 2^n T(0) + (2^n - 1)
<br> - T(n) = O(2^n)<br>
<br> (b) T(n) = 2T(n/2) + n<br>
<br> Visualize the Recursion Tree:
<br> - The root node represents T(n).
<br> - The root splits into two nodes, each representing T(n/2), plus n for the root level work.
<br> - Each T(n/2) further splits into two T(n/4) nodes, and so on.<br>
<br> Work Done at Each Level:
<br> - Level 0: T(n)
<br> - Level 1: 2T(n/2) + n
<br> - Level 2: 4T(n/4) + 2n
<br> - Level k: 2^k T(n/2^k) + k n<br>
<br> Summing the Work:
<br> - The number of nodes at level k is 2^k.
<br> - The total work at level k is 2^k T(n/2^k).
<br> - The total work for all levels is the sum of the work at each level.<br>
<br> Base Case:
<br> - T(1) is a constant.<br>
<br> Generalize:
<br> - T(n) = n + n + n + ... + n (log n times)
<br> - T(n) = O(n log n)
