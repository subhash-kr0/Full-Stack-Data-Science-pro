# Regression Analysis and Machine Learning Algorithms

## Regression Analysis

**Definition:**
Regression analysis is a statistical method used to model and analyze the relationship between a dependent variable and one or more independent variables. It helps in predicting the value of the dependent variable based on the values of the independent variables.

## Linear vs. Nonlinear Regression

**Linear Regression:**
- Models the relationship between variables as a straight line.
- Assumes a linear relationship between the dependent and independent variables.
- **Example:** \( y = \beta_0 + \beta_1 x \)

**Nonlinear Regression:**
- Models the relationship between variables as a curve.
- Does not assume a linear relationship.
- **Example:** \( y = \beta_0 + \beta_1 x + \beta_2 x^2 \)

## Simple Linear Regression vs. Multiple Linear Regression

**Simple Linear Regression:**
- Involves one independent variable and one dependent variable.
- **Equation:** \( y = \beta_0 + \beta_1 x \)

**Multiple Linear Regression:**
- Involves two or more independent variables.
- **Equation:** \( y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \ldots + \beta_n x_n \)

## Evaluating Regression Model Performance

**Common Metrics:**
- **Mean Absolute Error (MAE):** Average of absolute errors between predicted and actual values.
- **Mean Squared Error (MSE):** Average of squared errors between predicted and actual values.
- **Root Mean Squared Error (RMSE):** Square root of MSE.
- **R-squared (RÂ²):** Proportion of variance explained by the model.

## Overfitting in Regression Models

**Definition:**
Overfitting occurs when a model learns the noise in the training data rather than the underlying pattern. This results in poor generalization to new, unseen data.

**Indicators:**
- High accuracy on training data.
- Poor performance on validation/test data.

## Logistic Regression

**Usage:**
Logistic regression is used for binary classification problems where the target variable has two classes.

**Differences from Linear Regression:**
- Logistic regression predicts probabilities of class membership.
- Uses the sigmoid function to map predictions to probabilities.

## Odds Ratio in Logistic Regression

**Concept:**
The odds ratio represents the ratio of the odds of an event occurring to the odds of it not occurring. In logistic regression, it is used to describe the relationship between the independent variables and the probability of the event.

## Sigmoid Function in Logistic Regression

**Definition:**
The sigmoid function maps any real-valued number to a value between 0 and 1. It is defined as:
\[ \sigma(x) = \frac{1}{1 + e^{-x}} \]

**Usage:**
Used to convert the linear output of a logistic regression model into a probability.

## Evaluating Logistic Regression Model Performance

**Common Metrics:**
- **Accuracy:** Proportion of correct predictions.
- **Precision:** Proportion of true positive predictions among all positive predictions.
- **Recall:** Proportion of true positive predictions among all actual positives.
- **F1 Score:** Harmonic mean of precision and recall.
- **ROC-AUC:** Area under the Receiver Operating Characteristic curve.

## Decision Tree

**Definition:**
A decision tree is a tree-like model used for classification and regression tasks. It splits the data into branches to make predictions.

## Decision Tree Predictions

**Process:**
1. Start at the root node.
2. Split data based on feature values.
3. Follow the branches to leaf nodes.
4. Make predictions based on the majority class or average value at the leaf nodes.

## Entropy in Decision Trees

**Definition:**
Entropy measures the impurity or randomness in a dataset. It is used to determine the best splits in a decision tree.

**Formula:**
\[ \text{Entropy}(S) = - \sum_{i=1}^{c} p_i \log_2(p_i) \]

## Pruning in Decision Trees

**Definition:**
Pruning is the process of removing parts of the tree that do not provide additional power in predicting target variables, to prevent overfitting.

## Handling Missing Values in Decision Trees

**Methods:**
- **Surrogate Splits:** Use alternative splits when primary splits encounter missing values.
- **Imputation:** Fill missing values with the most frequent or average value.

## Support Vector Machine (SVM)

**Definition:**
SVM is a supervised learning algorithm used for classification and regression tasks. It finds the hyperplane that best separates the classes in the feature space.

## Margin in SVM

**Concept:**
The margin is the distance between the hyperplane and the closest data points from each class. SVM aims to maximize this margin.

## Support Vectors in SVM

**Definition:**
Support vectors are the data points closest to the hyperplane. They are critical in defining the position and orientation of the hyperplane.

## Handling Non-linearly Separable Data in SVM

**Methods:**
- **Kernel Trick:** Use kernel functions (e.g., RBF, polynomial) to transform data into higher dimensions where it becomes linearly separable.

## Advantages of SVM

**Advantages:**
- Effective in high-dimensional spaces.
- Works well with a clear margin of separation.
- Robust to overfitting, especially in high-dimensional spaces.

**Limitations:**
- Not suitable for large datasets due to high computational cost.
- Less effective with noisy data and overlapping classes.
