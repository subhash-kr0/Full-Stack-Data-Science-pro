```python
#bubble sorting

def bubbleSort(arr):
  for i in range(len(arr)-1, 0 , -1):
    for j in range(i):
      if(arr[j]>arr[j+1]):
        arr[j],arr[j+1] = arr[j+1],arr[j]
```

```python
arr = [5,4,3,2,1]
bubbleSort(arr)
print(arr)
```

```python
def bubbleSortOptimized(arr):
  for i in range(len(arr)-1, 0 , -1):
    isSorted = True
    for j in range(i):
      if(arr[j]>arr[j+1]):
        isSorted = False
        arr[j],arr[j+1] = arr[j+1],arr[j]

    if isSorted :
      print("Array is already sorted")
      break
```

```python
arr = [1,2,3,5,4]
bubbleSortOptimized(arr)
print(arr)
```

```python
# Given an array, check if the array is sorted ?

def isSorted(arr):
  isSorted = True
  for i in range(len(arr)-1):
    if arr[i]>arr[i+1]:
      isSorted = False
      break
  if isSorted :
    print("Array is sorted")
  else :
    print("Array is not sorted")
```

```python
#Selection Sorting
def selectionSort(arr):
  for i in range(len(arr)):
    min =i
    for j in range(i+1,len(arr)):
      if arr[j] <arr[min]:
        min =j
    #Find the smallest values
    arr[i],arr[min] = arr[min],arr[i]
```

```python
def insertion_sort(arr):
  for i in range(1,len(arr)):
    v = arr[i]
    j=i
    while(j>=1 and arr[j-1]>v):
      arr[j]=arr[j-1]
      j-=1
    arr[j]=v
```

```python
def search(arr, num):
  for i in range(len(arr)):
    if arr[i]==num:
      return i
  return -1

```

```python
def binarySearch(arr, num):
  left = 0
  right = len(arr) -1


  while(left <=right):
    mid = int(left + (right-left)/2)
    if arr[mid] == num:
      return mid
    elif arr[mid] > num :
      right = mid-1
    else :
      left = mid+1
  return -1


```

```python
def find_min(arr, sI):
  #Base condition
  if sI >=len(arr):
    return 99999999

  return min(arr[sI], find_min(arr,sI+1))


print(find_min([3,2,5],0))
```

```python
def moves(n, s ,d , a):
  # Base condition
  if n == 1 :
    return 1

  # move n-1 s->a , then move 1 biggest coin to s->d and then n-1 coin a-d
  return moves(n-1,s,a,d) + 1 + moves(n-1,a,d,s)

print(moves(1,'s','d','a'))
print(moves(2,'s','d','a'))
print(moves(3,'s','d','a'))
print(moves(200,'s','d','a'))
```

```python
def search_rec(arr, sI, data):
  #Base condition
  if sI >= len(arr):
    return -1
  if arr[sI] == data :
    return sI
  else :
    return search_rec(arr, sI+1 , data)

print( search_rec([4,5,3,1,2],0,7))
```

```python
def binarySearch(arr, left, right, num):
  #Base condition
  if left > right :
    return -1

  mid = int( left + (right-left)/2)

  if arr[mid] == num :
    return mid

  elif arr[mid] > num :
    return binarySearch(arr, left, mid-1, num)
  else :
    return binarySearch(arr, mid+1, right, num)


print(binarySearch([2,4,6,8,10],0,5,8))
print(binarySearch([2,4,6,8,10],0,5,7))
```

```python
def reverseArr(arr):
  i =0
  j = len(arr)-1

  while(i<j):
    #swap elements at i and j index
    arr[i],arr[j] = arr[j],arr[i]
    i +=1
    j -=1

arr = [4,5,1,2,3]
reverseArr(arr)
print(arr)
```

```python
def reverse_rec(arr, sI, eI):
  if sI >= eI :
    return
  arr[sI],arr[eI] = arr[eI],arr[sI] #swapping values
  reverse_rec(arr,sI+1,eI-1)

arr = [1,2,3,4,5]
reverse_rec(arr, 0, len(arr)-1)
print(arr)



```

```python
def rotate_k_times(arr,k) :
  k = k % len(arr)

  #reverse the entire array
  reverse_rec(arr,0,len(arr)-1)

  #reverse the firsr k elements
  reverse_rec(arr,0,k-1)

  #reverse the last n-k elements
  reverse_rec(arr,k,len(arr)-1)

arr = [1,2,3,4,5]
rotate_k_times(arr,2)
print(arr)
```

```python
def isPalindrome(arr):
  # 2 pointer approach
  left =0
  right = len(arr)-1

  while(left <right):
    if arr[left] != arr[right] :
      return False
    left +=1
    right -=1
  return True

arr = [ 1,2,3,3,2,1]
arr1 = [1,2,4]
arr3 = [1]
arr4 = [1,2,1,1]

print(isPalindrome(arr))
print(isPalindrome(arr1))
print(isPalindrome(arr3))
print(isPalindrome(arr4))
```

```python
def rotate(arr,k):
  k = k%len(arr)
  #Num of times the rotation has to take place
  for i in range(k) :
    #Logic to rotate the array
    prev = arr[0]
    curr = arr[0]
    arr[0] = arr[len(arr)-1]
    j=1
    while(j<len(arr)):
      curr = arr[j]
      arr[j] = prev
      prev = curr
      j+=1

arr1 = [1,2,3,4,5]
arr2 = [1,2,3,4,5]
arr3 = [1,2,3,4,5]

rotate(arr1,1)
print(arr1)
rotate(arr2,2)
print(arr2)
rotate(arr3,3)
print(arr3)
```
